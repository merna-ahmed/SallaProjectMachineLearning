```python
#Time Series Forecast
import pandas as pd

df['Date'] = pd.to_datetime(df['Date'])

df = df.sort_values('Date')

monthly_df = df.groupby(pd.Grouper(key='Date', freq='M')).sum().reset_index()

ts = monthly_df[['Date', 'Forecasted Calls']]
```

```python
pip install Prophet 
```

```python
from prophet import Prophet

prophet_df = ts.rename(columns={'Date': 'ds', 'Forecasted Calls': 'y'})

model = Prophet()
model.fit(prophet_df)

future = model.make_future_dataframe(periods=3, freq='M')  
forecast = model.predict(future)

forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(6)
```

```python
model.plot(forecast)
plt.title=("Forecasted Calls (Time Series Forecast)")
plt.show()
```

```python
#The forecast shows an upward trend in forecasted calls.
#The blue line represents the predicted values.
#The shaded area indicates the uncertainty interval.
#The wide range suggests higher uncertainty due to limited data.
```
